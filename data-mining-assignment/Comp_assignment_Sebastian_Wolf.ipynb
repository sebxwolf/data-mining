# Report to the mayor of London: Traffic accidents and risk factors

The mayor of London has asked for a report on road safety in the City of London. She/he wants the report to include answers to the following questions:
    - Is road safety improving or getting worse?
    - Who is most affected?
    - Which areas and which times do we see the most accidents?
    
To answer the mayor's questions, I prepared a Jupyter notebook studying data about accidents in the UK from the years 2005-2015.
    
## Loading, and preparing the data for analysis    

```python
# read the datasets
accidents = pd.read_csv("./data/Data_FULL-2005-2015/Accidents_2005-2015_categorical.csv")
casualties = pd.read_csv("./data/Data_FULL-2005-2015/Casualties_2005-2015_categorical.csv")
vehicles = pd.read_csv("./data/Data_FULL-2005-2015/Vehicles_2005-2015_categorical.csv")

# merge the datasets
accidents_casualties = pd.merge(accidents, casualties, on='Accident_Index')
accidents_vehicles = pd.merge(accidents, vehicles, on='Accident_Index')
vehicles_casualties = pd.merge(casualties, vehicles, on='Accident_Index')

# makes a date-time object 
accidents['DateTime'] = pd.to_datetime(accidents.Date + " " +  accidents.Time)

# sets the date-time object as the index of our dataframe
accidents.index = accidents['DateTime']

# creates a dataframe containing only accidents that were recorded by the London police forces: the Metropolitan Police and the City of London Police Forece
accidents_ldn = accidents[(accidents.Police_Force=="Metropolitan Police") | (accidents.Police_Force=="City of London")]

# creates a dataframe containing accidents recorded by police forces in the rest of the country
accidents_not_ldn = accidents[(accidents.Police_Force!="Metropolitan Police") & (accidents.Police_Force!="City of London")]

# creates a casualty dataframe containing only accidents that were recorded by the London police forces: the Metropolitan Police and the City of London Police Forece
accidents_casualties_ldn = accidents_casualties[(accidents_casualties.Police_Force=="Metropolitan Police") | (accidents_casualties.Police_Force=="City of London")]

# creates a vehicles dataframe containing accidents recorded by police forces in the rest of the country
accidents_vehicles_ldn = accidents_vehicles[(accidents_vehicles.Police_Force=="Metropolitan Police") | (accidents_vehicles.Police_Force=="City of London")]
```

## Question 1: Has the number of accidents in London gone down?

I first put the number of accidents in London into context - how has the number evolved and how does it compare to the rest of the country and to the size of the population?

#### Question 1 Analysis

```python
### PLOTS
# Number of monthly accidents, casualties, and vehicles involved graphed over time for London
fig, fig1 = plt.subplots(1, 1)
fig1.plot(accidents_ldn[['Number_of_Vehicles', 'Number_of_Casualties', 'Accident_Index']].resample('m').agg({'Number_of_Vehicles':'sum', 'Number_of_Casualties':'sum', 'Accident_Index':'count'}))
fig1.legend(('Number of Vehicles', 'Number of Casualties', 'Number of Accidents'))
fig1.set_ylabel("Monthly occurences")
fig1.set_xlabel("Year")
fig1.set_title('Plot 1: London')
plt.show()

# Number of monthly accidents, casualties, and vehicles involved graphed over time for the rest of the Country
fig, fig1 = plt.subplots(1, 1)
fig1.plot(accidents_not_ldn[['Number_of_Vehicles', 'Number_of_Casualties', 'Accident_Index']].resample('m').agg({'Number_of_Vehicles':'sum', 'Number_of_Casualties':'sum', 'Accident_Index':'count'}))
fig1.legend(('Number of Vehicles', 'Number of Casualties', 'Number of Accidents'))
fig1.set_ylabel("Monthly occurences")
fig1.set_xlabel("Year")
fig1.set_title('Plot 2:Rest of the Country')
plt.show()

# Number of monthly accidents in London vs the rest of the country
fig, fig2 = plt.subplots(1, 1)
fig2.plot(accidents_not_ldn[['Accident_Index']].resample('M').agg({'Accident_Index':'count'}),label='Rest of the Country')
fig2.plot(accidents_ldn[['Accident_Index']].resample('M').agg({'Accident_Index':'count'}),label='London')
fig2.set_ylabel("Monthly occurences")
fig2.set_xlabel("Year")
fig2.set_title('Plot 3: Accidents in London vs the rest of the country')
fig2.legend()
plt.show()

### TABLE
# Goal of table: Shows the absolute number of accidents, casualties and vehicles in London vs the rest side by side
# Resample dataframes to annual frequency
yearly_accidents_ldn = accidents_ldn[['Accident_Index','Number_of_Casualties','Number_of_Vehicles']].resample('a').agg({'Accident_Index':'count','Number_of_Vehicles':'sum', 'Number_of_Casualties':'sum'})
yearly_accidents = accidents_not_ldn[['Accident_Index','Number_of_Casualties','Number_of_Vehicles']].resample('a').agg({'Accident_Index':'count','Number_of_Vehicles':'sum', 'Number_of_Casualties':'sum'})

# change column names to faciliate merging of the two dataframes
yearly_accidents_ldn.columns = ['Accidents in London', 'Casualties in London', 'Vehicles involved in London']
yearly_accidents.columns = ['Accidents (UK except London)', 'Casualties (UK except London)', 'Vehicles involved (UK except London)']

# merge the two dataframes and change column order
combine = [yearly_accidents_ldn,yearly_accidents]
combines_yearly_accidents = pd.concat(combine,axis=1)
cols = ['Accidents in London', 'Accidents (UK except London)', 'Casualties in London', 'Casualties (UK except London)', 'Vehicles involved in London', 'Vehicles involved (UK except London)']
combines_yearly_accidents = combines_yearly_accidents[cols]

# change index
combines_yearly_accidents.index = combines_yearly_accidents.index.year
combines_yearly_accidents.index.rename('Year', inplace=True)

# introduce multi-hierarchical column headers
index = [['Accidents','Accidents','Vehicles','Vehicles','Casualties','Casualties'],['London','Rest','London','Rest','London','Rest']]
combines_yearly_accidents.columns = index
endpoints = pd.concat([combines_yearly_accidents.iloc[:1,:],combines_yearly_accidents.iloc[-1:,:]],axis=0)
endpoints
```

## Question 1 Takeaways:

- Plot 1 above shows that the number of accidents, casualties and vehicles **involved fluctuates a lot within years, but has remained somewhat stable over the 10 years overall.**
- It shows a downward trend for the years 2005-10, but in the time 2010-15 it has risen back to 2005 levels.
- This stands in contrast to the trend we see in Plot 2 for the rest of the country, where all three metrics show a clear downward trend.
- Over ten years, the number of **accidents per year in London has decreased from 26,812 to 25,232 whereas it has decreased from 171,894 to 114,806 in the Rest of the country, a very large difference!**
- This poses the question - what is the reason that accidents have decreased in the rest of the country but remained stable in London? 

## Question 2: Did accidents increase relative to the population?

Question 1 underlined the need to look at accident numbers in relative terms. In this section, we compare the number of accidents over time to the population growth.

#### Question 2 Analysis

```python
# import and reshape population data
population = pd.read_excel("./data/Data_FULL-2005-2015/ons-mye-population-totals.xls")

population.index= population['Year']
population = population.drop(columns='Year')
population.columns = ['Pop_LDN','Pop_GB']

# merge population data to yearly number of accidents
combine = [combines_yearly_accidents,population]
accidents_population = pd.concat(combine,axis=1)

# compute relative number of accidents
accidents_population['Pop_Rest'] = accidents_population['Pop_GB']-accidents_population['Pop_LDN']
accidents_population['Accidents/Pop in London'] = accidents_population[('Accidents', 'London')]/accidents_population['Pop_LDN']
accidents_population['Accidents/Pop Rest of UK'] = accidents_population[('Accidents', 'Rest')]/accidents_population['Pop_Rest']

### PLOT
accidents_population['Accidents/Pop in London'].plot()
accidents_population['Accidents/Pop Rest of UK'].plot()
plt.legend()
plt.title('Plot 4: Accidents in relative terms')

### TABLE
endpoints = pd.concat([accidents_population.iloc[:1,-2:]*100,accidents_population.iloc[-1:,-2:]*100],axis=0)
endpoints
```

## Question 2 Takeaways:

- Plot 2 above shows that **the number of accidents per inhabitant in London per year has decerased from 0.36% to 0.29% in the period 2005-15.**
- In the same time, it has **decreased from 0.34% to 0.21% in the rest of the country.**
- While the number of accidents in London has decreased relative to the size of population, improvements have been larger in the rest of the country.

## Question 3: Did accidents increase relative to the kilometers travelled?

Another 'driver' of the number of traffic accidents is the amount of kilometers travelled by traffic participants. The more people use road transport, the more accidents we expect. Therefore, I now put the number of traffic accidents into context with the number of kilometers travelled.

#### Question 3 Analysis

```python
# import and reshape road usage data
kilometers = pd.read_excel("./data/Data_FULL-2005-2015/kilometers_import.xlsx")

kilometers['Year'] = pd.to_datetime(kilometers.Year,format='%Y')
kilometers.index = kilometers['Year']
kilometers.index = kilometers.index.year
kilometers = kilometers.drop(columns='Year')

# create data frame that allows subtracting road usage in London from the rest of the GB to get the road usage for the 'rest of the country'
kilometers_ldn = kilometers[(kilometers.Region == 'London')].drop(columns='Region')
kilometers_gb = kilometers[(kilometers.Region == 'Great Britain')].drop(columns='Region')
kilometers_not_ldn = kilometers[(kilometers.Region == 'Great Britain')].drop(columns='Region').sub(kilometers[(kilometers.Region == 'London')]).drop(columns='Region')
kilometers_ldn = kilometers_ldn.reindex(sorted(kilometers_ldn.columns), axis=1)

# merge road usage data to yearly number of accidents
accidents_kilometers_temp = pd.merge(combines_yearly_accidents, kilometers_ldn, on='Year')

# re-label the columns
index = pd.Series(accidents_kilometers_temp.columns[-8:])
new_index = pd.Series(['Accidents_Ldn','Accidents_Rest','Vehicles_London','Vehicles_Rest','Casualties_Ldn','Casualties_Rest'])
index_temp = pd.Series(['LDN'] * 8)
new_index1 = index[-8:] + "_" + index_temp
index_temp = pd.Series(['Rest'] * 8)
new_index2 = index[-8:] + "_" + index_temp
merged_index = pd.concat([new_index, new_index1, new_index2], axis=0)

# cast the dataframe for analysis using re-labelled columns and merged data
accidents_kilometers = pd.merge(accidents_kilometers_temp, kilometers_not_ldn, on='Year')         
accidents_kilometers.columns = merged_index

# compute number of accidents relative to the number of kilometers travelled
accidents_kilometers['Accidents/bn motor vehicle km in London'] = accidents_kilometers['Accidents_Ldn']/accidents_kilometers['All motor vehicles_LDN']
accidents_kilometers['Accidents/bn motor vehicle km in Rest of UK'] = accidents_kilometers['Accidents_Ldn']/accidents_kilometers['All motor vehicles_Rest']

### PLOTS
accidents_kilometers['Accidents/bn motor vehicle km in London'].plot()
accidents_kilometers['Accidents/bn motor vehicle km in Rest of UK'].plot()
plt.legend()
plt.title('Plot 5: Number of accidents per bn km travelled')

### TABLE
# This table compares the endpoints of the timeseries
endpoints = pd.concat([accidents_kilometers.iloc[:1,-2:],accidents_kilometers.iloc[-1:,-2:]],axis=0)
endpoints
```

## Question 3 Takeaways:

- **Relative to road usage, the number of accidents in London has actually increased!** It rose from 853 to 864 accidents yearly per bn km travelled by motor vehicles over the time period 2005-15.
- In the rest of the country, this number has decreased from 58 to 52.
- Plot 5 reveals that these levels have remained relatively stable year-on-year.

## Question 7: What are risk factors for pedestrians?

Having studied the demographics of victims, I now look at potential risk factors. In which situations do most accidents occur? In this report I zero in on pedestrians and cyclists, because in relative terms, they are the most vulnerable. I start with studying the situation for pedestrians.

#### Question 7 Analysis

```python
### PLOTS
# heatmap for pedestrian casualties by time of day and day of the week
accidents_casualties_ldn.index = accidents_casualties_ldn['DateTime']
accidents_ldn_day_time = accidents_casualties_ldn[(accidents_casualties_ldn.Casualty_Class=='Pedestrian')]
accidents_ldn_day_time = accidents_ldn_day_time.groupby(['Day_of_Week',accidents_ldn_day_time.index.hour]).agg({'Number_of_Casualties':'sum'})
accidents_ldn_day_time = pd.pivot_table(accidents_ldn_day_time,values='Number_of_Casualties',index='Day_of_Week',columns='DateTime')
accidents_ldn_day_time = accidents_ldn_day_time.loc[['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'],:]
fig, ax = plt.subplots()
fig.set_size_inches(20, 5)
ax.set_title('Plot 14: Number of pedestrian casualties by time of day and day of the week')
sns.heatmap(accidents_ldn_day_time,linewidths=.2, cmap="Blues")
```

```python
# heatmap for fatal pedestrian casualties by time of day and day of the week
accidents_ldn_day_time = accidents_casualties_ldn[(accidents_casualties_ldn.Casualty_Severity=='Fatal') & (accidents_casualties_ldn.Casualty_Class=='Pedestrian')]
accidents_ldn_day_time = accidents_ldn_day_time.groupby(['Day_of_Week',accidents_ldn_day_time.index.hour]).agg({'Number_of_Casualties':'sum'})
accidents_ldn_day_time = pd.pivot_table(accidents_ldn_day_time,values='Number_of_Casualties',index='Day_of_Week',columns='DateTime')
accidents_ldn_day_time = accidents_ldn_day_time.loc[['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'],:]
fig, ax = plt.subplots()
fig.set_size_inches(20, 5)
ax.set_title('Plot 15: Number of pedestrian fatalities by time of day and day of the week')
sns.heatmap(accidents_ldn_day_time,linewidths=.2, cmap="Blues")
```

```python
# heatmap for pedestrian casualties by pedestrians location and crossing facilities present
accidents_ldn_day_time = accidents_casualties_ldn[(accidents_casualties_ldn.Casualty_Class=='Pedestrian')]
accidents_ldn_day_time = accidents_ldn_day_time.groupby(["Pedestrian_Crossing-Physical_Facilities",'Pedestrian_Location']).agg({'Number_of_Casualties':'sum'})
accidents_ldn_day_time = pd.pivot_table(accidents_ldn_day_time,values='Number_of_Casualties',index="Pedestrian_Location",columns='Pedestrian_Crossing-Physical_Facilities')
fig, ax = plt.subplots()
fig.set_size_inches(20, 5)
ax.set_title('Plot 16: Pedestrian casualties by pedestrians location and crossing facilities present')
sns.heatmap(accidents_ldn_day_time,linewidths=.2, cmap="Blues",vmax=10000)
```

```python
# heatmap for pedestrian casualties by pedestrians movement at the time of the accident and age band of casualty
heatmap = accidents_casualties_ldn[(accidents_casualties_ldn.Casualty_Class=='Pedestrian')].groupby(["Pedestrian_Location","Pedestrian_Movement"]).agg({"Casualty_Type":"count"})
heatmap = pd.pivot_table(heatmap,values='Casualty_Type',index='Pedestrian_Movement',columns='Pedestrian_Location')
fig, ax = plt.subplots()
fig.set_size_inches(20, 5)
ax.set_title('Plot 17: Pedestrian casualties by pedestrians movement at the time of the accident and age band of casualty')
sns.heatmap(heatmap,linewidths=.2, cmap="Blues", vmax=10000)
```

```python
# heatmap for pedestrian casualties by pedestrians location and age band of casualty
heatmap = accidents_casualties_ldn[(accidents_casualties_ldn.Casualty_Class=='Pedestrian')].groupby(["Age_Band_of_Casualty","Pedestrian_Location"]).agg({"Casualty_Type":"count"})
heatmap = pd.pivot_table(heatmap,values='Casualty_Type',index='Pedestrian_Location',columns='Age_Band_of_Casualty')
fig, ax = plt.subplots()
fig.set_size_inches(20, 5)
ax.set_title('Plot 18: Pedestrian casualties by pedestrians location and crossing facilities present')
sns.heatmap(heatmap,linewidths=.2, cmap="Blues")
```
